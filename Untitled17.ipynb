```python
import pycountry

# Load your aviation data (if not already loaded)
# df = pd.read_csv("aviation_data_updated (1).csv")

# Add country name column
def iso3_to_country_name(iso3):
    try:
        return pycountry.countries.get(alpha_3=iso3).name
    except:
        return None

df['country_name'] = df['iso3_country'].apply(iso3_to_country_name)

# Load population CSV (contains: Country Name, Country Code (ISO3), Year, Value)
pop_df = pd.read_csv("population.csv")

# Filter to the latest year
latest_year = pop_df['Year'].max()
pop_df = pop_df[pop_df['Year'] == latest_year]

# Rename and prepare for merging
pop_df = pop_df.rename(columns={
    'Country Code': 'iso3_country',
    'Value': 'population'
})

# Merge with original df on iso3_country
df = df.merge(pop_df[['iso3_country', 'population']], on='iso3_country', how='left')

# Preview
print(df[['iso3_country', 'country_name', 'population']].drop_duplicates().head(10))
```
